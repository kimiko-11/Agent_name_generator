```python

def get_iou(bb1, bb2):
  # assuring for proper dimension.
    assert bb1['x1'] < bb1['x2']
    assert bb1['y1'] < bb1['y2']
    assert bb2['x1'] < bb2['x2']
    assert bb2['y1'] < bb2['y2']
  # calculating dimension of common area between these two boxes.
    x_left = max(bb1['x1'], bb2['x1'])
    y_top = max(bb1['y1'], bb2['y1'])
    x_right = min(bb1['x2'], bb2['x2'])
    y_bottom = min(bb1['y2'], bb2['y2'])
  # if there is no overlap output 0 as intersection area is zero.
    if x_right < x_left or y_bottom < y_top:
        return 0.0
  # calculating intersection area.
    intersection_area = (x_right - x_left) * (y_bottom - y_top)
  # individual areas of both these bounding boxes.
    bb1_area = (bb1['x2'] - bb1['x1']) * (bb1['y2'] - bb1['y1'])
    bb2_area = (bb2['x2'] - bb2['x1']) * (bb2['y2'] - bb2['y1'])
  # union area = area of bb1_+ area of bb2 - intersection of bb1 and bb2.
    iou = intersection_area / float(bb1_area + bb2_area - intersection_area)
    assert iou >= 0.0
    assert iou <= 1.0
    return iou
```

```python

# At the end of below code we will have our train data in these lists
train_images=[]
train_labels=[]
```

```python
import os
output_folder = "/content/drive/MyDrive/RCNN_crops"
pos_folder = os.path.join(output_folder, "pos")
neg_folder = os.path.join(output_folder, "neg")
os.makedirs(pos_folder, exist_ok=True)
os.makedirs(neg_folder, exist_ok=True)

for e,i in enumerate(os.listdir(annot)):
    try:
        if i.startswith("airplane"):
            filename = i.split(".")[0]+".jpg"
            print(e, filename)
            image = cv2.imread(os.path.join(path, filename))
            df = pd.read_csv(os.path.join(annot, i))
            gtvalues=[]
            for index, row in df.iterrows():
                x1 = int(row.iloc[0].split(" ")[0])
                y1 = int(row.iloc[0].split(" ")[1])
                x2 = int(row.iloc[0].split(" ")[2])
                y2 = int(row.iloc[0].split(" ")[3])
                gtvalues.append({"x1":x1,"x2":x2,"y1":y1,"y2":y2})

                # Save positive ground-truth crops
                timage = image[y1:y2, x1:x2]  # correct order
                resized = cv2.resize(timage, (224,224))
                cv2.imwrite(os.path.join(pos_folder, f"{filename}_pos_{index}.jpg"), resized)

            ss.setBaseImage(image)
            ss.switchToSelectiveSearchFast()
            ssresults = ss.process()
            counter = 0
            falsecounter = 0
            flag = 0
            for e,result in enumerate(ssresults):
                if e >= 2000 or flag == 1:
                    break
                for gtval in gtvalues:
                    x, y, w, h = result
                    iou = get_iou(gtval, {"x1":x, "x2":x+w, "y1":y, "y2":y+h})
                    if falsecounter < 5 and iou < 0.3:
                        timage = image[y:y+h, x:x+w]  # correct order
                        resized = cv2.resize(timage, (224,224))
                        cv2.imwrite(os.path.join(neg_folder, f"{filename}_neg_{falsecounter}.jpg"), resized)
                        falsecounter += 1
                    if falsecounter >= 5:
                        flag = 1
                        break
    except Exception as e:
        print(e)
        print("error in "+filename)
        continue
```

```python

# conversion of train data into arrays for further training
X_new = np.array(train_images)
Y_new = np.array(train_labels)
```

```python

from keras.layers import Dense
from keras import Model
from keras import optimizers
```

```python
vgg = tf.keras.applications.vgg16.VGG16(include_top=True, weights='imagenet', input_tensor=None, input_shape=None, pooling=None, classes=1000)
for layer in vgg.layers[:-2]:
  layer.trainable = False
x = vgg.get_layer('fc2')
last_output =  x.output
x = tf.keras.layers.Dense(1,activation = 'sigmoid')(last_output)
model = tf.keras.Model(vgg.input,x)
model.compile(optimizer = "adam",
              loss = 'binary_crossentropy',
              metrics = ['acc'])
```

```python

model.summary()
model.fit(X_new,Y_new,batch_size = 64,epochs = 3, verbose = 1,validation_split=.05,shuffle = True)
```

```python
import matplotlib.pyplot as plt

def plot_training_history(history, save_folder="/content/drive/MyDrive/RCNN_plots"):
    import os
    os.makedirs(save_folder, exist_ok=True)

    # Plot Loss
    plt.figure(figsize=(8,6))
    plt.plot(history.history['loss'], label='Train Loss')
    if 'val_loss' in history.history:
        plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title("Model Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    loss_path = os.path.join(save_folder, "loss_curve.png")
    plt.savefig(loss_path)
    plt.show()
    print(f"Loss curve saved to: {loss_path}")

    # Plot Accuracy
    if 'accuracy' in history.history or 'acc' in history.history:
        plt.figure(figsize=(8,6))
        acc_key = 'accuracy' if 'accuracy' in history.history else 'acc'
        plt.plot(history.history[acc_key], label='Train Accuracy')
        if 'val_' + acc_key in history.history:
            plt.plot(history.history['val_' + acc_key], label='Validation Accuracy')
        plt.title("Model Accuracy")
        plt.xlabel("Epoch")
        plt.ylabel("Accuracy")
        plt.legend()
        acc_path = os.path.join(save_folder, "accuracy_curve.png")
        plt.savefig(acc_path)
        plt.show()
        print(f"Accuracy curve saved to: {acc_path}")

# Example usage:
# plot_training_history(hist_final)
```
